```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
aprocimacoes = 0.0

listaAprocimacoes = []
valorPi = []
x = []

for n in range(1000):
    aprocimacoes += (-1.0)**n/(2.0*n+1.0)
    print(aprocimacoes*4)
    listaAprocimacoes.append(aprocimacoes*4)
    valorPi.append(np.pi)
    x.append(n)
    
```

```python
plt.title("Aproximação de PI")
plt.plot(x, listaAprocimacoes,color='#17a589', label='Aproximações')
plt.plot(x, valorPi,color='red', label='ValorPI')

plt.legend()
```
